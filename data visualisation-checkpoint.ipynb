```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
mobile = pd.read_csv('train.csv')
```

```python
mobile
```

```python
mobile.head()
```

```python
mobile.rename_axis('S. No',axis='rows',inplace=True)
```

```python
mobile
```

```python
plt.figure(figsize=(12,6))
plt.title("Line chart")
sns.lineplot(data=mobile[0:11])
plt.xlabel('Mobiles')
           
```

```python
plt.figure(figsize=(12,6))
plt.title("Line chart")
sns.lineplot(data=mobile['battery_power'][0:11])
plt.xlabel('Mobiles')
           
```

```python
plt.figure(figsize=(20,6))
plt.title("Line chart")
sns.barplot(data=mobile[0:11])
plt.xlabel('Mobiles')
           
```

```python
plt.figure(figsize=(20,6))
plt.title("Line chart")
sns.barplot(x=mobile.index[0:11],y=mobile['battery_power'][0:11])
plt.xlabel('Mobiles')
           
```

```python
plt.figure(figsize=(20,6))
plt.title("Heatmap")
sns.heatmap(data=mobile[0:11])
plt.xlabel('Mobiles')
           
```

```python
plt.figure(figsize=(20,6))
plt.title("Heatmap")
sns.heatmap(data=mobile[0:11],annot=True)
plt.xlabel('Mobiles')
           
           
```

```python
plt.figure(figsize=(20,6))
plt.title("Scatter Plot")
sns.scatterplot(x=mobile.index[0:100],y=mobile['battery_power'][0:100])
plt.xlabel('Mobiles')
           
```

```python
plt.figure(figsize=(20,6))
plt.title("Scatter Plot")
sns.regplot(x=mobile.index[0:100],y=mobile['battery_power'][0:100])
plt.xlabel('Mobiles')
           
```

```python
plt.figure(figsize=(20,6))
plt.title("Scatter Plot")
sns.scatterplot(x=mobile.index[0:100],y=mobile['battery_power'][0:100],hue=mobile['mobile_wt'][0:100])
plt.xlabel('Mobiles')
           
```

```python
plt.figure(figsize=(20,6))
plt.title("Scatter Plot")
sns.lmplot(x='clock_speed'[0:100],y='battery_power'[0:100],hue='mobile_wt'[0:100],data=mobile)
plt.xlabel('Mobiles')
```

```python
sns.swarmplot(x=mobile.index[0:10],y=mobile['blue'][0:10])
```
